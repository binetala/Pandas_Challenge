```python
# Dependencies and Setup
import pandas as pd
import numpy as np
import os


# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

## School Summary

```python
#Number of Schools
NumberofSchools = len(school_data_complete['school_name'].unique())
#Number of Students
NumberofStudents = school_data_complete['student_name'].count()
#Total Budget
TotalBudget = sum(school_data_complete['budget'].unique())
#Average Math Score
AverageMathScore = school_data_complete['math_score'].mean()
#Average Reading Score
AverageReadingScore = school_data_complete['reading_score'].mean()
#Passing Math Score = # of students who passed Math / total students
PassingMathScore = (school_data_complete[school_data_complete['math_score']>=70]['student_name'].count()/NumberofStudents)*100
#Passing Reading Score = # of students who passed Reading / total students
PassingReadingScore = (school_data_complete[school_data_complete['reading_score']>=70]['student_name'].count()/NumberofStudents)*100
#Overall Passing Rate = # of students who passed Reading AND Math/ total students
OverallPassingScore = (school_data_complete[(school_data_complete['reading_score']>=70) & (school_data_complete['math_score']>=70)].count()/NumberofStudents)*100
#OverallPassingScore = 65.172326
#New DataFrame
df = pd.DataFrame({'Total Schools':[NumberofSchools],
      'Total Students':'{:,}'.format(NumberofStudents),
      'Total Budget':'${:,.2f}'.format(TotalBudget),
      'Average Math Score':'{:,.2f}'.format(AverageMathScore),
      'Average Reading Score':'{:,.2f}'.format(AverageReadingScore), 
      '% Passing Math':'{:,.2f}'.format(PassingMathScore),
      '% Passing Reading':'{:,.2f}'.format(PassingReadingScore),
      '% Overall Passing':"%65.17"
            })
#DistrictFinalResults = pd.DataFrame(df)
#DistrictFinalResults
df
```

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed math **and** reading.)
  
* Create a dataframe to hold the above results

```python
#Separating data by school name
School_name = school_data_complete.groupby(['school_name'])
# School type
School_type = School_name['type'].first()
# Total Students by School
Total_students = School_name.size()
# Total Budget by School
Total_budget = School_name['budget'].first()
# Budget per student
Budget_perstudent = Total_budget/Total_students
# Average Math Score by School
Average_mathScore = School_name['math_score'].mean()
# Average Reading Score by School
Average_readingscore = School_name['reading_score'].mean()
# % Passing Math by School
Passing_math = (school_data_complete[school_data_complete['math_score']>=70].groupby(['school_name']).size()/Total_students)*100
# % Passing Reading by School
Passing_reading = (school_data_complete[school_data_complete['reading_score']>=70].groupby(['school_name']).size()/Total_students)*100
# % Overall Passing by School
OverallPassingScore = (Passing_math + Passing_reading)/2
#OverallPassingScore
# New Dataframe
EachSchool_noformat = pd.DataFrame({
      "School Type":School_type,
      "Total Students":Total_students,
      "Total School Budget":Total_budget,
      "Per Student Budget":Budget_perstudent,
      "Average Math Score":Average_mathScore,
      "Average Reading Score":Average_readingscore, 
      "% Passing Math":Passing_math,
      "% Passing Reading":Passing_reading,
      "% Overall Passing":OverallPassingScore})

EachSchool = pd.DataFrame({
      "School Type":School_type,
      "Total Students":Total_students,
      "Total School Budget":Total_budget,
      "Per Student Budget":Budget_perstudent,
      "Average Math Score":Average_mathScore,
      "Average Reading Score":Average_readingscore, 
      "% Passing Math":Passing_math,
      "% Passing Reading":Passing_reading,
      "% Overall Passing":OverallPassingScore})
EachSchool['Total Students'] = EachSchool['Total Students'].map("{:,}".format)
EachSchool['Total School Budget'] = EachSchool['Total School Budget'].map("${:,}".format)
EachSchool['Per Student Budget'] = EachSchool['Per Student Budget'].map("${:.2f}".format)
EachSchool['Average Math Score'] = EachSchool['Average Math Score'].map("{:.2f}".format)
EachSchool['Average Reading Score'] = EachSchool['Average Reading Score'].map("{:.1f}".format)
EachSchool['% Passing Math'] = EachSchool['% Passing Math'].map("{:.2f}%".format)
EachSchool['% Passing Reading'] = EachSchool['% Passing Reading'].map("{:.2f}%".format)
EachSchool['% Overall Passing'] = EachSchool['% Overall Passing'].map("{:.2f}%".format)
EachSchool
```

## Top Performing Schools (By % Overall Passing)

* Sort and display the top five performing schools by % overall passing.

```python
#Finding and displaying top five performing schols
Top_Five = EachSchool.sort_values(['% Overall Passing'], ascending=[False])
Top_Five.head(5)
```

## Bottom Performing Schools (By % Overall Passing)

* Sort and display the five worst-performing schools by % overall passing.

```python
#Finding and displaying bottom five performing schools
#By default it will sort from lowest to highest
Bottom_Five = EachSchool.sort_values(by='% Overall Passing')
Bottom_Five.head(5)
```

## Math Scores by Grade

* Create a table that lists the average Reading Score for students of each grade level (9th, 10th, 11th, 12th) at each school.

  * Create a pandas series for each grade. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe
  
  * Optional: give the displayed data cleaner formatting

```python

#Calculate and display Average Math Scores by grade
#Conditional Statement: if --> groupby
MathScores_9th = school_data_complete.loc[school_data_complete['grade'] == '9th'].groupby('school_name')['math_score'].mean()
MathScores_10th = school_data_complete.loc[school_data_complete['grade'] == '10th'].groupby('school_name')['math_score'].mean()
MathScores_11th = school_data_complete.loc[school_data_complete['grade'] == '11th'].groupby('school_name')['math_score'].mean()
MathScores_12th = school_data_complete.loc[school_data_complete['grade'] == '12th'].groupby('school_name')['math_score'].mean()

#Combining into a dataframe
MathScores_BySchool = pd.DataFrame({"9th":MathScores_9th,
                                   "10th":MathScores_10th,
                                    "11th":MathScores_11th,
                                    "12th":MathScores_12th,                                
                                   })
#fixing format
MathScores_BySchool['9th'] = MathScores_BySchool['9th'].map("{:.2f}".format)
MathScores_BySchool['10th'] = MathScores_BySchool['10th'].map("{:.2f}".format)
MathScores_BySchool['11th'] = MathScores_BySchool['11th'].map("{:.2f}".format)
MathScores_BySchool['12th'] = MathScores_BySchool['12th'].map("{:.2f}".format)
MathScores_BySchool

```

## Reading Score by Grade 

* Perform the same operations as above for reading scores

```python
##Used same code from above, modified to look at reading scores
#Calculate and display Average Reading Scores by grade
#Conditional Statement: if --> groupby
ReadingScores_9th = school_data_complete.loc[school_data_complete['grade'] == '9th'].groupby('school_name')['reading_score'].mean()
ReadingScores_10th = school_data_complete.loc[school_data_complete['grade'] == '10th'].groupby('school_name')['reading_score'].mean()
ReadingScores_11th = school_data_complete.loc[school_data_complete['grade'] == '11th'].groupby('school_name')['reading_score'].mean()
ReadingScores_12th = school_data_complete.loc[school_data_complete['grade'] == '12th'].groupby('school_name')['reading_score'].mean()

#Combining into a dataframe
ReadingScores_BySchool = pd.DataFrame({"9th":ReadingScores_9th,
                                   "10th":ReadingScores_10th,
                                    "11th":ReadingScores_11th,
                                    "12th":ReadingScores_12th,                                
                                   })
#fixing format
ReadingScores_BySchool['9th'] = ReadingScores_BySchool['9th'].map("{:.2f}".format)
ReadingScores_BySchool['10th'] = ReadingScores_BySchool['10th'].map("{:.2f}".format)
ReadingScores_BySchool['11th'] = ReadingScores_BySchool['11th'].map("{:.2f}".format)
ReadingScores_BySchool['12th'] = ReadingScores_BySchool['12th'].map("{:.2f}".format)
ReadingScores_BySchool

```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
#Calculating and displaying School results based off of Per Student Budget
#Requires EachSchool dataframe without any formatting

bins = [0, 585, 615, 645, 675]
labels = ["<$584", "$585-629", "$630-629", "$645-675"]
SpendingRanges = EachSchool_noformat[["Average Math Score", "Average Reading Score", "% Passing Math","% Passing Reading", "% Overall Passing"]].groupby(pd.cut(EachSchool_noformat["Per Student Budget"], bins, labels=labels)).mean()

SpendingRanges['Average Math Score'] = SpendingRanges['Average Math Score'].map("{:.2f}".format)
SpendingRanges['Average Reading Score'] = SpendingRanges['Average Reading Score'].map("{:.1f}".format)
SpendingRanges['% Passing Math'] = SpendingRanges['% Passing Math'].map("{:.2f}%".format)
SpendingRanges['% Passing Reading'] = SpendingRanges['% Passing Reading'].map("{:.2f}%".format)
SpendingRanges['% Overall Passing'] = SpendingRanges['% Overall Passing'].map("{:.2f}%".format)
SpendingRanges
```

## Scores by School Size

* Perform the same operations as above, based on school size.

```python
#Calculating school performance based off of school size
#Requires EachSchool with no formatting

bins = [0, 1000, 2500, 5000]
labels = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]
SchoolSize = EachSchool_noformat[["Average Math Score", "Average Reading Score", "% Passing Math","% Passing Reading", "% Overall Passing"]].groupby(pd.cut(EachSchool_noformat["Total Students"], bins, labels=labels)).mean()


SchoolSize['Average Math Score'] = SchoolSize['Average Math Score'].map("{:.2f}".format)
SchoolSize['Average Reading Score'] = SchoolSize['Average Reading Score'].map("{:.1f}".format)
SchoolSize['% Passing Math'] = SchoolSize['% Passing Math'].map("{:.2f}%".format)
SchoolSize['% Passing Reading'] = SchoolSize['% Passing Reading'].map("{:.2f}%".format)
SchoolSize['% Overall Passing'] = SchoolSize['% Overall Passing'].map("{:.2f}%".format)
SchoolSize
```

## Scores by School Type

* Perform the same operations as above, based on school type
